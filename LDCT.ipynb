```python
sums, sums_squared = 0, 0

for c, SOPInstance_UID in enumerate(tqdm(labels.SOPInstanceUID)):
    StudyInstance_UID = labels.StudyInstanceUID.iloc[c]
    SeriesInstance_UID = labels.SeriesInstanceUID.iloc[c]
    SOPInstance_UID = labels.SOPInstanceUID.iloc[c]
    dcm_path = ROOT_PATH/StudyInstance_UID/SeriesInstance_UID/SOPInstance_UID
    dcm_path = dcm_path.with_suffix(".dcm")
    
    dcm = pydicom.read_file(dcm_path).pixel_array / 255
    
    dcm_array = cv2.resize(dcm, (224,224))
    
    mean = 0
    std = 0
    noise = np.random.normal(mean, std, dcm_array.shape)
    
    image_G_noisy =  dcm_array + noise
    
    image_G_noisy[image_G_noisy < 0] = 1e-6
    
    image_G_noisy[np.isnan(image_G_noisy)] = 1e-6
    
    lambda_ = 25
    
    noise = np.random.poisson(lambda_, size=dcm_array.shape)
    
    image_P_noisy = image_G_noisy + noise
    
    label = labels.pe_present_on_image.iloc[c]
    train_or_val = "train" if c<80000 else "val"
    current_save_path = SAVE_PATH/train_or_val/str(label)/SeriesInstance_UID
    current_save_path.mkdir(parents=True, exist_ok=True)
    np.save(current_save_path/SOPInstance_UID, image_P_noisy)
    
    normalizer = 224*224
    if train_or_val == "train":
        sums += np.sum(dcm_array) / normalizer
        sums_squared += (dcm_array **2).sum() /normalizer
```

```python
def load_file(path):
    return np.load(path).astype(np.float32)
```

```python
train_transforms = transforms.Compose([
                    transforms.ToTensor(),
                    transforms.Normalize(1.50,3.20),
                    transforms.RandomAffine(degrees=(-5,5), translate=(0,0.05), scale=(0.9,1.1)), 
                    transforms.RandomResizedCrop((224, 224), scale=(0.15,1))
])
```

```python
train_dataset = torchvision.datasets.DatasetFolder("LDCT2/train/0/",loader=load_file, extensions="npy",transform=train_transforms)
```

```python
fig, axis = plt.subplots(2,2, figsize=(9,9))
for i in range(2):
    for j in range(2):
        random_index = np.random.randint(0,80000)
        x_ray, label = train_dataset[random_index]
        axis[i][j].imshow(x_ray[0], cmap="gray")
        axis[i][j].set_title(label)
```

```python
mean = sums / 80000
std = np.sqrt((sums_squared / 80000) - mean**2)
```

```python
batch_size = 16
num_workers = 4

train_loader = torch.utils.data.DataLoader(train_dataset,batch_size=batch_size, num_workers=num_workers, shuffle=True)
```
